# Split Categories into Multiple Rows

This notebook splits rows that have multiple categories in a single column, creating a new row for each category.

```python
import pandas as pd
```

```python
# Load your data
# Replace with your actual file path
df = pd.read_excel('Ecocert_final.xlsx')  # or pd.read_csv('your_file.csv')

# Display initial data
print("Original DataFrame:")
print(df.head())
print(f"\nShape: {df.shape}")
```

```python
# Define the column that contains multiple categories
# Change 'specuation' to your actual column name
CATEGORY_COLUMN = 'specuation'

# Check what separator is used in your categories (comma, semicolon, etc.)
print("Sample values in the category column:")
print(df[CATEGORY_COLUMN].head(10))
```

```python
# Function to split categories and create new rows
def split_categories_to_rows(df, category_column, separator=','):
    """
    Split rows with multiple categories into separate rows.
    
    Parameters:
    - df: Input DataFrame
    - category_column: Name of the column with multiple categories
    - separator: Character used to separate categories (default: comma)
    
    Returns:
    - DataFrame with one category per row
    """
    # Strip whitespace and split by separator
    df[category_column] = df[category_column].astype(str)
    
    # Split the categories and explode into separate rows
    df_split = df.assign(
        **{category_column: df[category_column].str.split(separator)}
    ).explode(category_column, ignore_index=True)
    
    # Strip whitespace from each category
    df_split[category_column] = df_split[category_column].str.strip()
    
    return df_split


# Apply the function
# Change separator if your data uses something other than comma
df_split = split_categories_to_rows(df, CATEGORY_COLUMN, separator=',')

print("\nDataFrame after splitting categories:")
print(df_split.head(20))
print(f"\nNew shape: {df_split.shape}")
```

```python
# Verify the split worked correctly
print("\nUnique categories after split:")
print(df_split[CATEGORY_COLUMN].value_counts())
```

```python
# Save the result
output_file = 'Ecocert_final_split.xlsx'
df_split.to_excel(output_file, index=False)
print(f"\nSaved to: {output_file}")
```

```python
# Example with sample data (for testing)
# This shows how the function works with example data

sample_data = {
    'id': [1, 2, 3],
    'name': ['Product A', 'Product B', 'Product C'],
    'specuation': ['cat1, cat2', 'cat3', 'cat4, cat5, cat6'],
    'value': [100, 200, 300]
}

df_sample = pd.DataFrame(sample_data)
print("Sample data BEFORE split:")
print(df_sample)

df_sample_split = split_categories_to_rows(df_sample, 'specuation', separator=',')
print("\nSample data AFTER split:")
print(df_sample_split)
```
